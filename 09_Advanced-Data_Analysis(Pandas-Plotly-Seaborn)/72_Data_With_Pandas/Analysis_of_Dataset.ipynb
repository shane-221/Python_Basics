```python
import pandas as pd





```

```python
# Loading the data
df = pd.read_csv("Salaries_By_College_Major.csv")
pd.options.display.float_format = '{:,.2f}'.format      ## Converts it into an easily readable format
df.head()
```

```python
# Initial checks

df.head()

        # Checking the naming conventions of columns
df.rename(columns ={ "Undergraduate Major": "Undergraduate_Major",
            "Starting Median Salary":"Starting_Median_Salary",
            "Mid-Career Median Salary":"Mid-Career_Median_Salary",
            "Mid-Career 10th Percentile Salary":"Mid-Career_10th_Percentile_Salary",
            "Mid-Career 90th Percentile Salary":"Mid-Career_90th_Percentile_Salary",}, inplace = True)

        # Checking the types of data
df.dtypes
df.astype({"Starting_Median_Salary":"float64" ,
           "Mid-Career_Median_Salary":"float64"})
```

```python
# checks for missing data
df.isna().sum()
        # Tells you the count of the missing values
df.dropna(inplace=True)
        # Checks for the shape of the data
df.shape
df.isnull().sum()
```

## Challenge Questions( Use of choosing columns  + Max, Min ,idxmax() etc)

---


Now that we've found the major with the highest starting salary, can you write the code to find the following:

What college major has the highest mid-career salary? How much do graduates with this major earn? (Mid-career is defined as having 10+ years of experience).

Which college major has the lowest starting salary and how much do graduates earn after university?

Which college major has the lowest mid-career salary and how much can people expect to earn with this degree?

```python
# Question 1:
highest_mid_career_salary = df["Mid-Career_10th_Percentile_Salary"].idxmax()
print("Question 1:")
display(df.loc[highest_mid_career_salary, ["Undergraduate_Major","Mid-Career_10th_Percentile_Salary" ]])


# Question 2:
lowest_starting_salary = df["Starting_Median_Salary"].idxmin()
print("Question 2:")
display(df.loc[lowest_starting_salary, ["Undergraduate_Major", "Starting_Median_Salary"]])

# Question 3:
lowest_mid_career_salary = df["Mid-Career_Median_Salary"].idxmin()
print("Question 3:")
display(df.loc[lowest_mid_career_salary, ["Undergraduate_Major", "Mid-Career_Median_Salary"]])
```

### Use of the insert function

```python
## Create a new column with a new function called spread
spread =df ["Mid-Career_90th_Percentile_Salary"]- df["Mid-Career_10th_Percentile_Salary"]
# Insert function parameters (Where column, name , column_Data)
df.insert(1, "Spread", spread)
print(df.head())
```

### Use of the Sort values function.

```python
## Wish to sort the Spread column in ascending or descending
## sort_values(name of the columns, whether you want it ascending or not )
df.sort_values("Spread", ascending = False)
```
