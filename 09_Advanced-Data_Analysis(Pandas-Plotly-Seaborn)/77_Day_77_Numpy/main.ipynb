### Accessing the dimensions

#### One dimensional array

````python
one_array = np.array([1.1 , 9.2, 8.1,4.7 ])
one_array.shape         # Seeing the rows because its a 1d block so only shows one number

one_array[2]            # Access the third item like before
one_array.ndim          # Helps to see the columns
````

#### Two dimensional array( Matrix)

````python
array_2d  = np.array([
                        [1, 2, 3, 9],
                        [5, 6, 7, 8]
                        ])                  # Behaves as a nested list

print(f'array_2d has {array_2d.ndim} dimensions')       # Columns
print(f'Its shape is {array_2d.shape}')                 # Rows
print(f'It has {array_2d.shape[0]} rows and {array_2d.shape[1]} columns')
print(array_2d[1,2])            # Third value in the second row
print(array_2d[0, :])           # O column and all the rows
````

## Broadcasting, Scalars and Matrix Multiplication

#### Adding vectors

````python
v1 = np.array([4, 5, 2, 7])
v2 = np.array([2, 1, 3, 3])
print(v1+v2)
                            # Numpy will add the vectors together
                            ## Subtraction and multiplication  works as well.
````

####  Multiplying with a scalary :Broadcasting

Now, oftentimes you'll want to do some sort of operation between an array and a single number. In mathematics, this single number is often called a scalar. For example, you might want to multiply every value in your NumPy array by 2. Numpy will broadcast ( make the scalr into an arrray thats compatible with the larger array)

````python
array_2d = np.array([[1, 2, 3, 4],
                     [5, 6, 7, 8]])

array_2d+ 10            ## will  broadcast then add value by value.
````

#### Multiplying with a vector
We follow the rules of linear algebra. The inner dimensions must match and the outer dimesions will be the shape. It will be the rows x columns 

````python
a1 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])

b1 = np.array([[4, 1, 3],
               [5, 8, 5]])

result = np.matmul(a1, b1)
print(result )
````

#### Manipulating Images as ndarrays

````python
from skimage import data
img = data.astronaut()
````

````python
#### What is the data type?
type(img)
print(img)          #### will give you the rgb values

img.shape       # In numpy images ita ( rows, columns, channels)
img.ndim
````

````python
#### Converting the image to black
sRGB  = img /255
grayscale = 0.2126*sRGB[:,:,0] + 0.7152*sRGB[:,:,1]+0.0722*sRGB[:,:,2]

plt.imshow(grayscale, cmap="gray")
````
