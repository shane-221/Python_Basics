#### One dimensional array

```python
one_array = np.array([1.1 , 9.2, 8.1,4.7 ])
one_array.shape         # Seeing the rows because its a 1d block so only shows one number

one_array[2]            # Access the third item like before
one_array.ndim          # Helps to see the columns
```

#### Two dimensional array( Matrix)

```python
array_2d  = np.array([
                        [1, 2, 3, 9],
                        [5, 6, 7, 8]
                        ])                  # Behaves as a nested list

print(f'array_2d has {array_2d.ndim} dimensions')       # Columns
print(f'Its shape is {array_2d.shape}')                 # Rows
print(f'It has {array_2d.shape[0]} rows and {array_2d.shape[1]} columns')
print(array_2d[1,2])            # Third value in the second row
print(array_2d[0, :])           # O column and all the rows
```

#### N dimensions
An array with n dimensions is known as a tensor.

```python
array_n = np.array([[[0, 1, 2, 3],
                           [4, 5, 6, 7]],

                         [[7, 86, 6, 98],
                          [5, 1, 0, 4]],

                          [[5, 36, 32, 48],
                           [97, 0, 27, 18]]])
```
