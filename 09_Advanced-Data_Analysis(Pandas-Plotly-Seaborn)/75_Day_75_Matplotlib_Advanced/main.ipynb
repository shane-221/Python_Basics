### Challenge

Plot the Tesla stock price against the Tesla search volume using a line chart and two different axes.

```python
import matplotlib.pyplot as plt

# Sizing the graph
plt.figure(figsize=(16,10), dpi =120)
plt.xticks(fontsize= 14 , rotation =90)
plt.yticks(fontsize=14 , rotation=90)


# Drawing the graph
ax1 = plt.gca()
ax2 = ax1.twinx()

    # Functions:
ax1.plot(tesla_df["MONTH"],
         tesla_df["TSLA_WEB_SEARCH"],
         linewidth =3,
         label = "TSLA WEB SEARCH",
         color ="b")
ax2.plot(tesla_df["MONTH"],
         tesla_df["TSLA_USD_CLOSE"],
         linewidth =3,
         label = "TSLA USD CLOSE",
         color ="r")                            # Can use hex codes/ standard ones.

# Functionalities
ax1.set_xlabel("Date")
ax1.set_ylabel("Tesla Web Search", color = "b")
ax2.set_ylabel("Tesla USD Close Price( USD)", color = "r")
plt.title("Tesla Web Search vs Price", fontsize=18)

fig = plt.gcf()
fig.legend(loc ="lower right")


# showing the graph-make sure to use plt instead of ax1
plt.show()
```

#### Adding tickmarks to the same graphs

```python
import matplotlib.pyplot as plt


# Sizing the graph
plt.figure(figsize=(16,10), dpi =120)
plt.xticks(fontsize= 14 , rotation =90)
plt.yticks(fontsize=14 , rotation=90)






# Drawing the graph
ax1 = plt.gca()
ax2 = ax1.twinx()

    # Functions:
ax1.plot(tesla_df["MONTH"],
         tesla_df["TSLA_WEB_SEARCH"],
         linewidth =3,
         label = "TSLA WEB SEARCH",
         color ="b")
ax2.plot(tesla_df["MONTH"],
         tesla_df["TSLA_USD_CLOSE"],
         linewidth =3,
         label = "TSLA USD CLOSE",
         color ="r")                            # Can use hex codes/ standard ones.

# Functionalities
    # General:
ax1.set_xlabel("Date")
ax1.set_ylabel("Tesla Web Search", color = "b")
ax2.set_ylabel("Tesla USD Close Price( USD)", color = "r")
plt.title("Tesla Web Search vs Price", fontsize=18)

 # Tickmarks
import matplotlib.dates as mdates       # Import module from the package
Years= mdates.YearLocator()             # Get a tick for every year
months= mdates.MonthLocator()           # Gets a tick fo the months
YearsFMT = mdates.DateFormatter('%Y')   # Displays the date labels ina  specific manner

ax1.xaxis.set_major_locator(Years)
ax1.xaxis.set_minor_locator(months)
ax1.xaxis.set_major_formatter(YearsFMT)

    # Legend
fig = plt.gcf()
fig.legend(loc ="lower right")


# showing the graph-make sure to use plt instead of ax1
plt.show()
```

```python
unemployment_df = pd.read_csv("./Data/UE Benefits Search vs UE Rate 2004-19.csv")
unemployment_df.head()
```

```python
# Convert the months into average yearly v
unemployment_df["MONTH"] = pd.to_datetime(unemployment_df["MONTH"])
```

```python

#=====================================================================Diagram===================================================================================#
import matplotlib.pyplot as plt
    # Sizing of the diagram
plt.figure(figsize = (16,10), dpi =120)
plt.xticks(fontsize= 14 , rotation = 45)
plt.yticks(fontsize= 14, rotation = 90)


    # adjsuting the dates
import matplotlib.dates as mdates
Years= mdates.YearLocator()             # Find the years
Months = mdates.MonthLocator()          # Find the months
YearFMT = mdates.DateFormatter('%Y')    # In this format


    #  Twin axis
ax1 =  plt.gca()
ax2 = ax1.twinx()

ax1.plot(unemployment_df["MONTH"],
         unemployment_df["UNRATE"],
         linewidth =3,
         label = "UE RATE",
         color= "r")
ax2.plot(unemployment_df["MONTH"],
         unemployment_df["UE_BENEFITS_WEB_SEARCH"],
         linewidth =3,
         label = "UE BENEFITS WEB SEARCH",
         color ="b")

ax1.xaxis.set_major_locator(Years)      # Major locator in years
ax1.xaxis.set_minor_locator(Months)     # Minor locator in months
ax1.xaxis.set_major_formatter(YearFMT)  # Years in this format.

    # Other functionalities
ax1.set_xlabel("Date")
plt.title("UE Rate vs UE Benefits", fontsize=18, loc ="center")
ax2.set_ylabel("UE Benefit", color = "b")
ax1.set_ylabel("FRED U/E Rate", color = "r")

        # Adding a grey grid to chart
ax1.grid(True, linestyle = "--", color = "grey")

    # Legend
function =plt.gcf()
function.legend(loc ="lower center")
```
